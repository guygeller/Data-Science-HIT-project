```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df  = pd.read_csv(r'C:\Users\avivr\OneDrive\Desktop\New folder\Marvel_Dataframe_after_drop_and_fill7.csv')
df2 = pd.read_csv(r'C:\Users\avivr\OneDrive\Desktop\New folder\Marvel_Dataframe_after_drop_and_fill8.csv')
gender_col = df['Gender'].tolist()

for i, gender in enumerate(gender_col):
    if gender != 'Male':
        if gender_col[i] == 'female':
            gender_col[i] = 'Female';
        elif gender != 'Female':
            gender_col[i] = 'other'

df['Gender'] = gender_col
df['Gender'].value_counts().plot(kind = 'pie')
```

```python

mstatus_labels = ['Married', 'Single']
mstatus = df['Marital Status'].tolist()
for i,x in enumerate(mstatus):
    if x not in mstatus_labels:
        mstatus[i] = 'Other'

df['Marital Status'] = mstatus
df['Marital Status'].value_counts().plot(kind = 'pie')
```

```python
citizen_labels = ['American']
citizen = df['Citenzenship'].tolist()
for i,x in enumerate(citizen):
    if x not in citizen_labels:
        citizen[i] = 'Other'
df['Citenzenship'] = citizen

df['Citenzenship'].value_counts().plot(kind = 'pie')
```

```python
reality_label = ['Earth-616']
reality = df['Reality'].tolist()
for i,x in enumerate(reality):
    if x not in reality_label:
        reality[i] = 'Other'
df['Reality'] = reality

df['Reality'].value_counts().plot(kind = 'pie')
```

```python
apearences_average = pd.DataFrame.mean(df['Appearnces'])

appearnces = df['Appearnces'].tolist()

for i,x in enumerate(appearnces):
    if x <= apearences_average:
        appearnces[i] = 'Under Average'
    else:
        appearnces[i] = 'Above Average'
            
df['Appearnces'] = appearnces
df['Appearnces'].value_counts().plot(kind = 'pie')

```

```python
df_test  = pd.read_csv(r'C:\Users\avivr\OneDrive\Desktop\New folder\Marvel_Dataframe_after_drop_and_fill7.csv')
apearences_average = pd.DataFrame.mean(df_test['Appearnces'])

appearnces = df_test['Appearnces'].tolist()
for i,x in enumerate(appearnces):
    if x <= 1:
        appearnces[i] = 'Low'
    elif x > 3 and x <= 15:
        appearnces[i] = 'Mid'
    else:
        appearnces[i] = 'High'
df_test['Appearnces'] = appearnces
df_test['Appearnces'].value_counts().plot(kind = 'pie')

mstatus_labels = ['Married', 'Single']
mstatus = df_test['Marital Status'].tolist()
for i,x in enumerate(mstatus):
    if x not in mstatus_labels:
        mstatus[i] = 'Other'

df_test['Marital Status'] = mstatus

```

```python
#mapping = {'Single':1, 'Married':2, 'Other':0}
#df_test.replace(mapping, inplace=True)
ct1 = pd.crosstab(df_test['Marital Status'], df_test['Appearnces'], normalize='index')
ct1.plot(kind='bar', figsize=(8,3))
plt.title('Marital Status VS Apearences', fontsize=10)
plt.xlabel('Status')
plt.ylabel('Apeareance')
```
